```python
# Data Processing and Exploration
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
from IPython import display

import scipy.stats as stats
```

```python
df_train_raw = pd.read_csv('Data/train.csv')
df_test_raw = pd.read_csv('Data/test.csv')
```

## SalePrice

```python
# 'SalePrice' summary
df_train_raw['SalePrice'].describe()
```

```python
# 'SalePrice' plots
sns.distplot(df_train_raw['SalePrice']).set_title('SalePrice')

plt.figure()
prob = stats.probplot(df_train_raw['SalePrice'], plot=plt)
```

```python
# Plot Log('SalePrice')
log_SalePrice = np.log(df_train_raw['SalePrice'])
sns.distplot(log_SalePrice)
plt.figure()
prob = stats.probplot(log_SalePrice, plot=plt)
```
